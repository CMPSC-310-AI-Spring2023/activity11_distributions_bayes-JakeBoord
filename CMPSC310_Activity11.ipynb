## Binomial Distribution

For the binomial distribution, the plot created has appears to hover around the 50 mark on the y-axis, with some variance in the outputs. They appear to bounce from the 40-60 mark. Given this is a binomial distribution with a 50% probability of "success," the following graph appears to follow this trend.

The graph appears fairly continuous, and it continues to hover around the midpoint.

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

Given the uniform distribution, here we witness a plot with various outputs from 1 to 10. Given uniform has an equal chance to produce any output, the graph is randomized with numbers between 1 and 10. There is no 1 number that has a greater chance than the other, and the plot displays just that.

While also continuous, it apppears to bounce much more than binomial.

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

The normal distribution shows a plot that contains most values between 0 and 0.2; however, it does appear that we have some outliers. There are some peaks that go beyond the peak.

The mean and standard deviation are necessary for the bands to appear equal on both sides (if cut in half and flipped on each other)

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print("Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))

# gnb.predict_proba(X_test)
```
